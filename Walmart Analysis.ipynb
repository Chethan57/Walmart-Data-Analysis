```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Walmart_df=pd.read_csv("BA Assignment Walmart Sales.csv")
Walmart_df
```

```python
last_col = 11
last_row=1000
Walmart_df=pd.read_csv("BA Assignment Walmart Sales.csv",usecols=range(last_col+1),nrows=last_row)
Walmart_df
```

```python
Walmart_df['Total Sales']=Walmart_df['Unit price'] *Walmart_df['Quantity']
Walmart_df
```

```python
Walmart_df.dtypes
```

```python
g1=Walmart_df.groupby(['City','Branch']
```

```python
total_Sales_Revenue=g1[['Unit price','Quantity']].sum()
total_Sales_Revenue['Total Sales']=total_Sales_Revenue['Unit price'] *total_Sales_Revenue['Quantity'] 
total_Sales_Revenue
```

```python
city_sales = Walmart_df.groupby(['City'])['Total Sales'].sum()
city_sales
```

```python
plt.figure(figsize=(6,4))
plt.bar(city_sales.index, city_sales.values)
plt.title('Total Sales Revenue by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.show()
```

```python
branch_sales = Walmart_df.groupby(['Branch'])['Total Sales'].sum()
branch_sales
```

```python
plt.figure(figsize=(6,4))
plt.bar(branch_sales.index, branch_sales.values)
plt.title('Total Sales Revenue by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()
```

```python
total_Sales_Revenue=total_Sales_Revenue.reset_index
total_Sales_Revenue
```

```python
Walmart_df['Total Revenue'] = Walmart_df['Quantity'] * Walmart_df['Unit price']
Walmart_df
```

```python
average_price_df = Walmart_df.groupby(['City', 'Branch'])['Total Revenue'].sum() / Walmart_df.groupby(['City', 'Branch'])['Quantity'].sum()
average_price_df = average_price_df.reset_index()
average_price_df.rename(columns={0: 'Average Price'}, inplace=True)
average_price_df
```

```python
Walmart_df['Date'] = pd.to_datetime(Walmart_df['Date'])
Walmart_df['Month'] = Walmart_df['Date'].dt.month
```

```python
monthwise_performance = Walmart_df.groupby(['Month', 'Product line', 'Gender', 'Payment'])[['Total Sales', 'Quantity']].sum().reset_index()
monthwise_performance
```

```python
plt.subplot(1, 2, 1)
monthwise_sales = monthwise_performance.groupby('Month')['Total Sales'].sum()
monthwise_sales.plot(kind='line', marker='x', color='g', label='Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales Month over Month')
plt.xticks(range(1, 5))  
plt.legend()
plt.show()
```

```python
monthwise_performance['Total Revenue']=monthwise_performance['Total Sales']* monthwise_performance['Quantity']
monthwise_performance
```

```python
plt.subplot(1, 2, 2)
monthwise_revenue = monthwise_performance.groupby('Month')['Total Revenue'].sum()
monthwise_revenue.plot(kind='line', marker='o', color='r', label='Total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Month over Month')
plt.xticks(range(1, 5))  
plt.legend()
plt.show()
```
